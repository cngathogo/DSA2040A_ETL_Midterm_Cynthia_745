```python

import pandas as pd

#Loading the files
raw_data = pd.read_csv("data/raw_data.csv")
incremental_data = pd.read_csv("data/incremental_data.csv")

# Previewing the datasets
print("Raw Data - Head:")
print(raw_data.head())
print("\nRaw Data - Info:")
print(raw_data.info())

print("\nIncremental Data - Head:")
print(incremental_data.head())
print("\nIncremental Data - Info:")
print(incremental_data.info())

# Checking for missing values, duplicates, and unusual columns
print("\nRaw Data - Missing Values:")
print(raw_data.isnull().sum())
print("\nRaw Data - Duplicates:", raw_data.duplicated().sum())

print("\nIncremental Data - Missing Values:")
print(incremental_data.isnull().sum())
print("\nIncremental Data - Duplicates:", incremental_data.duplicated().sum())

```
